```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# EXAMPLE 2

```python
def f(x):
    x1 = x[0]
    x2 = x[1]
    val = x1**4 + x2**4 - 2*x1*x2
    return val
```

```python
# Generate grid
N = 100
box = 2.
x1 = np.linspace(-box, box, N)
x2 = np.linspace(-box, box, N)

x1_grid, x2_grid = np.meshgrid(x1, x2)
f_grid = f([x1_grid, x2_grid])
```

```python
plt.figure(figsize=(9, 4.5))
#plt.contourf(x1, x2, f_grid, levels=np.linspace(-0.1, 1.01, 25))
plt.contourf(x1, x2, f_grid, levels=10)
plt.colorbar()

plt.xlabel('X1', size=14)
plt.xticks(size=14)

plt.ylabel('X2', size=14)
plt.yticks(size=14)

plt.xlim(-0.1, 0.5)
plt.ylim(-1.1, 0)

plt.xlim(-box, box)
plt.ylim(-box, box)

#plt.legend(loc='best', fontsize=14)

plt.tight_layout()
plt.savefig(fname="plots/optimization_example_2.pdf", format="pdf")
plt.show()
```
